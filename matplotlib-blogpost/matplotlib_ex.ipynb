```python
import numpy as np
import matplotlib.pyplot as plt

# Generating two random arrays of floats
xs = np.random.randn(20)

ys = np.random.randn(20)

# Creating the pyplot, "co": cyan and circles
plt.figure(figsize=(9, 5), dpi = 100)
plt.plot(xs, ys, "co")
plt.show()
```

```python
# Generating two random arrays of floats
x1s = np.random.randn(20)
x2s = np.random.randn(20)

y1s = np.random.randn(20)
y2s = np.random.randn(20)

# Creating the pyplot, 
# "co": cyan and circles, "r^" red and triangles 

plt.figure(figsize=(9, 5), dpi = 100)
plt.plot(x1s, y1s, "co")
plt.plot(x2s, y2s, "r^")
plt.show()
```

```python
plt.figure(figsize=(9, 5), dpi = 100)
plt.plot(x1s, y1s, "co", label = "cyan dots")
plt.plot(x2s, y2s, "r^", label = "red pyramids")

# Adding more artists
plt.title("My Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

```python
# Initializing a figure and axes object
fig, ax1 = plt.subplots()

# Adjusting the figure parameters
fig.set_size_inches(9, 5)
fig.set_dpi(100)

# Plotting the data
ax1.plot(x1s, y1s, "co", label = "cyan dots")
ax1.plot(x2s, y2s, "r^", label = "red pyramids")

# Adding more artists
ax1.set_title("My Plot")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

plt.show()
```

```python
# Initializing a figure and axes object
fig, ax1 = plt.subplots()

# Adjusting the figure parameters
fig.set_size_inches(9, 5)
fig.set_dpi(100)

# Plotting the data
ax1.plot(x1s, y1s, "co", label = "cyan dots")

# Creating a second axes with the same x-axis as our first axes
ax2 = ax1.twinx()

ax2.plot(x2s, y2s, "r^", label = "red pyramids")

# Adding artists
ax1.set_title("My Plot")
ax1.set_xlabel("x")

# Here we set different y-labels for each axes or subplot
ax1.set_ylabel("dot ys", color = "blue")
ax2.set_ylabel("pyramid ys", color = "red")

plt.show()
```

```python
# Creating a figure and an array of axes objects that
# is one row by two columns
fig, axs = plt.subplots(1, 2)

# Adjusting the figure parameters
fig.set_size_inches(9, 5)
fig.set_dpi(100)

# Plotting the data

# for our first subplot
axs[0].plot(x1s, y1s, "co", label = "cyan dots")

# and our second subplot
axs[1].plot(x2s, y2s, "r^", label = "red pyramids")

# Adding artists to our first subplot
axs[0].set_title("My Cyan Plot")
axs[0].set_xlabel("x")
axs[0].set_ylabel("dot ys", color = "blue")

# Adding artists to our second subplot
axs[1].set_title("My Red Plot")
axs[1].set_xlabel("x")
axs[1].set_ylabel("pyramid ys", color = "red")

plt.show()
```
